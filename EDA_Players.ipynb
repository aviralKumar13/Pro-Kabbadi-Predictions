```python
player_df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Pro-Kabbadi All Seasons Data/All_Players_All_Seasons_Latest.csv')
```

```python
player_merged_data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Pro-Kabbadi All Seasons Data/merge_it.csv')
```

```python
player_df.head()
```

```python
player_all_data = pd.merge(player_df,player_merged_data, on=['team_name','season'])
```

```python
df = player_all_data.copy()
```

```python
player_all_data.describe()
```

```python
player_all_data['succ_raid_percent']=(player_all_data['success_raids']/(player_all_data['success_raids']+player_all_data['unsuccess_raids']))*100
player_all_data['succ_tackle_percent'] = (player_all_data['success_tackles']/player_all_data['tackles'])*100
```

```python
team_all_data['spt'] = (team_all_data['team_super_raid'] + team_all_data['team_super_tackles'] + team_all_data['team_all_outs_inflicted']) - team_all_data['team_all_outs_conceded']
```

```python
player_all_data['succ_raid_percent'].fillna(0,inplace=True)
player_all_data['succ_tackle_percent'].fillna(0,inplace=True)
```

```python
player_all_data.head()
```

```python

def player_plot_per_season(df,season):
    df1 = df[df.season==season]
    df1 = df1[df1.success_raids > 100]
    df1 = df1.sort_values(by=['succ_raid_percent'],ascending=False).head(10)
    fig = go.Figure(data=[    
    go.Bar(name='succ_raid_percent', x=df1.name, y=df1.succ_raid_percent),
    go.Bar(name='team_total_points_rank', x=df1.name, y=df1.team_total_points_rank)
    ])
    fig.update_layout(showlegend=True,title_text='Season : '+str(season) ,title=dict(x=0.5,y=0.9))
#     fig.update_layout(legend_orientation="h",legend=dict(x=0, y=-0.2))
    fig.update_layout(barmode='group')
    fig.show()
```

```python
def team_tackle_plot_per_season(df,season):
    df1 = df[df.season==season]
    df1 = df1.sort_values(by=['team_successful_tackles'],ascending=False).head(10)
    fig = go.Figure(data=[    
    go.Bar(name='team_successful_tackles', x=df1.team_name, y=df1.team_successful_tackles),
    go.Bar(name='team_total_points_rank', x=df1.team_name, y=df1.team_total_points_rank)
    ])
    fig.update_layout(showlegend=True,title_text='Season : '+str(season) ,title=dict(x=0.5,y=0.9))
#     fig.update_layout(legend_orientation="h",legend=dict(x=0, y=-0.2))
    fig.update_layout(barmode='group')
    fig.show()
```

```python
def team_raid_plot_per_season(df,season):
    df1 = df[df.season==season]
    df1 = df1.sort_values(by=['team_successful_raids'],ascending=False).head(10)
    fig = go.Figure(data=[    
    go.Bar(name='team_successful_raids', x=df1.team_name, y=df1.team_successful_raids),
    go.Bar(name='team_total_points_rank', x=df1.team_name, y=df1.team_total_points_rank)
    ])
    fig.update_layout(showlegend=True,title_text='Season : '+str(season) ,title=dict(x=0.5,y=0.9))
#     fig.update_layout(legend_orientation="h",legend=dict(x=0, y=-0.2))
    fig.update_layout(barmode='group')
    fig.show()
```

```python
def team_player_plot_success_raids(df,season):
    df1 = df[df.season==season]
    #df1 = df1.sort_values(by=['team_successful_raids'],ascending=False)
    fig = go.Figure(data=[    
    #go.Bar(name='player_successful_raids', x=df1.name, y=df1.success_raids),
    go.Bar(name='team_name', x=df1.team_name, y=df1.success_raids,
             text=(df1.name),textposition='auto')
    ])
    fig.update_layout(showlegend=True,title_text='Season : '+str(season) ,title=dict(x=0.5,y=0.9))
#     fig.update_layout(legend_orientation="h",legend=dict(x=0, y=-0.2))
    fig.update_layout(barmode='stack',xaxis={'categoryorder':'array', 'categoryarray':['Dabang Delhi K.C.','Bengal Warriors','Bengaluru Bulls','Haryana Steelers']})
    fig.show()
```
