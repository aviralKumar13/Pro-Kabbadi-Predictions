```python
import pandas as pd
import numpy as np
```

```python
seasons = [1,2,3,4,8,26,49]
names = ['Team All-outs Conceded','Team All-outs Inflicted','Team Average Raid Points','Team Average Tackle Points','Team Avg Points Scored','Team DOD Raid Points','Team Raid Points','Team Successful Raids','Team Successful Tackles','Team Super Raid','Team Super Tackles','Team Tackle Points','Team Total Points Conceded','Team Total Points']
```

```python
len(names)
```

```python
for i in names:
    df = pd.DataFrame()
    for j in seasons:
        df1 =  pd.read_csv(str(i)+"_"+str(j)+".csv")
        df1[i] = df1['value']
        df1[i+'_rank'] = df1['rank']
        df1.drop(['value'],axis=1,inplace=True)
        df1.drop(['rank'],axis=1,inplace=True)
        df1['season']=j
        df = df.append(df1,ignore_index=True)
        df.to_csv(i+'.csv',index=False)
```
